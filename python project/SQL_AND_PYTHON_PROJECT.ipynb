```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('/content/workbook.xlsx')
```

```python
data.head()
```

```python
df = data.drop(columns=["ID"])
df.head()
```

PEOPLE BUY BIKE

```python
ax = sns.countplot(data = df , x = 'Purchased Bike')
ax.bar_label(ax.containers[0])
```

Income vs. Purchased Bike

```python
sns.lineplot(x='Purchased Bike',y='Income',data= df)
```

Age vs. Purchased Bike

```python
sns.lineplot(x='Purchased Bike', y='Age Group', hue='Gender', data=df)
plt.bar_label(ax.containers[0])
```

Commute Distance vs. Purchased Bike

```python

sns.boxplot(x='Purchased Bike', y='Commute Distance', data=df)
```
